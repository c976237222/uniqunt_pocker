```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from poker_ai.range import Range
```

```python
# 
action_list = ['raise0','raise1', 'raise2', 'raise3', 'call1', 'call2', 'call3']
range_dir = 'range/GTO/'
pos = 'UTG/'
for action in action_list:
    filename = range_dir + pos + action + '.txt'
    r = Range(filename, name=filename)
    ratio = r.count_ratio()
    print(action, '{:.1f}%'.format(ratio*100))
    r.plot(filename)
```
